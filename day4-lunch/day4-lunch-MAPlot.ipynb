#male 14A female 14A to measure the male/female differences during the #same time point of development

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Read in expression file
df_fpkms = pd.read_csv( "../../qbb2021/data/fpkms.csv", index_col="t_name" )
df_fpkms
```

```python
#Initialize male and female columns

df_male_14A = df_fpkms.loc[:,"male_14A"]
df_female_14A = df_fpkms.loc[:,"female_14A"]
```

```python
#Remove the header
x = df_male_14A[1:]
y = df_female_14A[1:]
```

```python
#Calculate log2 fold change for each gene; male is numerator
#Found equation on wikipedia
m = np.log2((x+1)/(y+1))
#calculate the log2 average expression for each gene in both the male
#and female expression levels 
a = 0.5*np.log2((x+1)*(y+1))
```

```python
#Plot the a (avg expression values) for x axis and m (log2 fold change
#expression values for the y axis)
#matplotlib plot function
ig, ax = plt.subplots()
ax.plot(a,m, '.')
plt.title("MA Plot: Male14A and Female14A")
plt.xlabel("Average Gene Expression")
plt.ylabel("Log2 Fold Change of Male to Female")
plt.show()
```

# Time Series

```python
#Read in replicate file
df_fpkms_replicate = pd.read_csv( "~/qbb2021/data/replicates.csv", index_col="t_name" )
#subset file by only sx1 gene
df_sxl_replicate = df_fpkms_replicate.loc["FBtr0331261",:]
#isolate only the expression values
df_sxl_only_fpkms_replicate = df_sxl_replicate


#Subset fpmks file by only Sxl gene
df_sxl = df_fpkms.loc["FBtr0331261",:]
df_sxl_only_fpkms = df_sxl[1:]
df_sxl_only_fpkms
```

```python
#Isolate only expression levels for Sxl
df_sxl = df_fpkms.loc["FBtr0331261",:]

#Isolate only the expression values (not index)
df_sxl_only_fpkms = df_sxl[1:]

#Set x axis values
x = ["10","11","12","13","14A","14B","14C","14D"]
#set replicate x axis value
x_rep = ["14A", "14B", "14C", "14D"]

#Set y axis values
y_male1 = df_sxl_only_fpkms.iloc[0:8]
y_female1 = df_sxl_only_fpkms.iloc[8:16]

#set y axis values for replicates
y_male_rep = df_sxl_only_fpkms_replicate.iloc[0:4]
y_female_rep = df_sxl_only_fpkms_replicate.iloc[4:8]

fig, ax = plt.subplots()
ax.scatter(x,y_male1, label="Male")
ax.scatter(x,y_female1, label="Female")
ax.scatter(x_rep, y_male_rep, marker="+", label="Male Replicate")
ax.scatter(x_rep, y_female_rep, marker="+", label="Female Replicate")
ax.set_title("Sxl Expression Over Time")
plt.xlabel("Developmental Stage")
plt.ylabel("Log2 Gene Expression")
ax.legend()
plt.show()
```
